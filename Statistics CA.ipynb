# Description of the data

    The database is composed by 1 dependent variable, price, and 15 independent variables 
    
    Description of the variables:
    price: the variable price is supposed to be the price of the house (including the lot). It ranges from 5000 to 775000 (we assume USD).
    lotsize: it ranges from 0 to 12200. Using the sort(D$lotSize) we see that the rows with 0 are 2. We assume that the rows with lotsize 0 are flats without land.
    Age: range from 0 to 225. Since the average is 27, we can assume the unit of measurement is years.
    LandValue: range from 200 to 412600. As for the price, we assume USD.
    Living area: range from 616 to 5228 
    pctCollege: no assumptions on the nature of the variable
    bedrooms: range from 1 to 7
    fireplaces: rance from 0 to 4
    bathrooms: range from 0 to 4.5 (maybe the half is due to the fact that in some of them there is only the shower.
    rooms: range from 2 to 12
    heating: possible values: electric, hot air, hot water/steam.
    fuel: electric, gas, oil
    sewer: none, pubblic/commercial, septic
    waterfront: yes, no
    newConstruction: yes, no
    centralAir: yes, no

## Forward Model

Start with no variables and try to add one-by-one the variable responsible for the largest R² value:

### Step 1

```r
ddum1<-ddum
```

```r
colnames(ddum1)[12] <- "heating_hot_air"
colnames(ddum1)[13] <- "heating_hot_water_steam"
summary(ddum1)
```

```r
attach(ddum1)
```
